```python
from nltk.tokenize import wordpunct_tokenize

def process_impression(impression):
    impression = impression.lower()
    return ' '.join(wordpunct_tokenize(impression))

df = pd.read_csv('test-data.csv')
df = df.applymap(process_impression)
df
```

```python
df.to_csv('test-data-proc.csv',index=False)
```

```python
from datasets import load_dataset
from torch.utils.data import Dataset, DataLoader

dataset = load_dataset("csv", data_files={"test": "/content/test-data-proc.csv"})#,"test":"/content/test-data.csv"}

```

```python
dataloader = DataLoader(dataset['test']['findings'], batch_size=16)
```

```python
import re
space_proc = lambda x: re.sub(" +"," ",x.replace("\n"," "))

max_source_length =  375
max_target_length =  241

def preprocess_function(sample,padding="max_length"):
    # add prefix to the input for t5
    inputs = [space_proc(item) for item in sample["findings"]]

    # tokenize inputs
    model_inputs = tokenizer(inputs, max_length=max_source_length, padding=padding, truncation=True)

    # Tokenize targets with the `text_target` keyword argument
    with tokenizer.as_target_tokenizer():
      labels = tokenizer(text_target=sample["impression"], max_length=max_target_length, padding=padding, truncation=True)

    # # If we are padding here, replace all tokenizer.pad_token_id in the labels by -100 when we want to ignore
    # # padding in the loss.
    if padding == "max_length":
        labels["input_ids"] = [
            [(l if l != tokenizer.pad_token_id else -100) for l in label] for label in labels["input_ids"]
        ]

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
model.to(device='cuda')
model.eval()
```

```python
import torch
from tqdm import tqdm
res = []
for batch in tqdm(dataloader):
  with torch.no_grad():
    inp = tokenizer(batch,max_length=375,padding="max_length",
                      truncation=True,return_tensors='pt').to(device='cuda')
    
    ops = model.generate(**inp).to(device='cpu')#,max_new_tokens=max_target_length
    
    ops2 = tokenizer.batch_decode(ops,skip_special_tokens=True)
    res.extend(ops2)
```

```python
# import pandas as pd

# df = pd.read_csv('/content/test-data.csv')
df['pred'] = res
df
```
